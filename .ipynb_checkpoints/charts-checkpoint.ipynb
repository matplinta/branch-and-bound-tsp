```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def time_plot(data):
    
    plt.figure(figsize=(12, 6))
    plt.xlabel("Liczba procesów")
    plt.ylabel("Czas [s]")
    plt.title("Czas [s]")
    plt.grid(True)
    
    time = data.apply(lambda x: x.time  ,axis=1)
    x = data['split']
    y = time
    plt.plot( x, y, marker='o')
```

```python
def time_plot_all(data):
    methodSet = set(data['split'])
    
    plt.figure(figsize=(12, 6))
    plt.xlabel("Liczba procesorów")
    plt.ylabel("Czas [s]")
    plt.title("Czas [s]")
    plt.grid(True)
    
    for method in methodSet:
        data_methoded = data.loc[data['split'] == method]
        time = data_methoded.apply(lambda x: x.time  ,axis=1)
        x = data_methoded['maxTaskPar']
        y = time
        plt.plot( x, y, marker='o', label="Split: {}".format(method))

    plt.legend(loc='best')
```

```python
def speedup_plot_all(data):
    methodSet = set(data['split'])
    
    plt.figure(figsize=(12, 6))
    plt.xlabel("Liczba procesorów")
    plt.ylabel("Przyśpieszenie")
    plt.title("Przyśpieszenie")
    plt.grid(True)
    
    for method in methodSet:
        data_methoded = data.loc[data['split'] == method]
        t_1 = data_methoded['time'].iloc[0]
        speedup = data_methoded.apply(lambda x: t_1 / x.time  ,axis=1)
        x = data_methoded['maxTaskPar']
        y = speedup
        plt.plot( x, y, marker='o', label="Split: {}".format(method))

    plt.legend(loc='best')
```

```python
def efficiency_plot_all(data):
    methodSet = set(data['split'])
    
    plt.figure(figsize=(12, 6))
    plt.xlabel("Liczba procesorów")
    plt.ylabel("Efektywność")
    plt.title("Efektywność")
    plt.grid(True)
    
    for method in methodSet:
        data_methoded = data.loc[data['split'] == method]
        t_1 = data_methoded['time'].iloc[0]
        efficiency = data_methoded.apply(lambda x: t_1 / (x.time *x.maxTaskPar) ,axis=1)
        x = data_methoded['maxTaskPar']
        y = efficiency
        plt.plot( x, y, marker='o', label="Split: {}".format(method))
    plt.legend(loc='best')
```

# Konfiguracja  - opis
Wykonano symulacje dla danych lokalizacji 13 miast, przedstawionych poniżej. Jako inicjalizacyjny węzeł przyjęto węzeł nr 1.  
Napisany program rozdziela ilość wszystkich możliwych poddrzew w miarę możliwości po równo na ilość procesów zdefiniowaną w parametrze konfiguracyjnym _split_.  
A więc przykładowo ilość wszystkich gałęzi dla 13 miast wynosi (13-1)! = 479001600. W tym sprawozdaniu przykładowo wybrano dwie ilości procesów, 4 oraz 10, definiujące odpowiednio szerokość poddrzewa na proces jako 119750400 oraz 47900160. Program rekurencyjnie sprawdza wszystkie poddrzewa po kolei, sprawdzając przebyty dystans i porównując go z minimalnym. W przypadku natrafienia na wartość większą niż zdefiniowaną w minimalnym - zostaje odcięta analizowana gałąź (branch&bound).

Innymi słowy, ilość tasków/procesów przekładała się na przyjętą szerokość poddrzew. Im większa ilość tasków, tym szerokość poddrzew była mniejsza per task, naturalnie.

Wszystkie przedstawione wyniki są rezultatem uśrednienia wartości 5 pomiarów.
Natomiast linie przedstawione na wykresach nie mają żadnej wartości formalnej a raczej tylko charakter merytoryczny i służą zwiększeniu czytelności wykresów.

# Znajdowanie optymalnej gruboziarnistości poddrzew 


## Basic - lokalne minimum per task

W przykładzie programu z lokalnymi minimami obserwujemy najlepsze wyniki dla szerokości poddrzew odpowiadającej 10 taskom/procesom. Kolejne zmniejszenie szerokości skutkuje pogorszeniem wyników.

```python
spectrum_basic = pd.read_csv("logs/basic_12cores_even_processes.csv")
time_plot(spectrum_basic)
```

# Speedup i Efficiency - local min

Poniżej przesdstawiono wykresy czasu, przyspieszenia oraz efektywności programu w funkcji ilości procesorów. Należy jednak zauważyć, iż dla przyjętych dwóch wielkości poddrzew (odpowiadającym ilości procesów 4 oraz 10), dane mają znaczenie/sens tylko dla ilości procesorów mniejszej niż wartości zdefiniowanej przez _split_, ponieważ zwiększenie ilości dostępnych procesorów powoduje nadmiarowy, niewykorzystany procesor, w związku z czym wartości dla kolejnych punktów pomiarowych przyjmują tę samą wartość, co obserwujemy na wykresach dla splitu 4 powyżej l. proc. 4 i adekwatnie dla splitu 10 dla ilości procesorów większej niż 10. 

```python
basic = pd.read_csv("logs/basic_4_10.csv")
time_plot_all(basic)
speedup_plot_all(basic)
efficiency_plot_all(basic)
```

## Advanced - wymiana minimum między taskami

W tym przypadku widzimy ciągłe polepszenie się wyników przy zmniejszaniu szerokości poddrzew per task. Optymalna wartość okazała się być najmniejszą wielkością poddrzew zmierzoną - tj. dla 12 tasków.

```python
spectrum_adv = pd.read_csv("logs/advanced_12cores_even_processes.csv")
time_plot(spectrum_adv)
```

# Speedup i Efficiency - global min

W tym przypadku obserwujemy maksymalne polepszenie po przekroczeniu ilości 3 procesorów, później wyniki się ustabilizowały dla obu wartości parametru _split_. Wynika to prawdopodobnie ze struktury danych - prawdopodobnie po 3 tasku znajdowane jest minimum globalne i pozostałe taski są bardzo szybko wycinane.

Podczas gdy metoda ta daje lepsze wyniki czasowe niż lokalne minima, metoda z globalnymi minimami gorzej się zrównolegla, tutaj osiągając maksymalne przyspieszenie wartości ~4 dla splitu 10, podczas gdy przy metodzie lokalnych minimum obserwowaliśmy przyspieszenie wartości ~7 dla tej samej wartości splitu = 10. Dla szerokości poddrzew odpowiadającej wartości split = 4 obserwujemy przyspieszenie podobnego rzętu ~3 dla obu metod wymiany minimum między procesami. 

```python
adv = pd.read_csv("logs/advanced_4_10.csv")
time_plot_all(adv)
speedup_plot_all(adv)
efficiency_plot_all(adv)
```
